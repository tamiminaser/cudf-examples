```python
%load_ext cudf.pandas

import pandas as pd
```

```python
raw_df = pd.read_csv("data_10000000.csv").head(3000000)
```

```python
%%timeit -n5 -r10
print(raw_df.groupby("country_of_birth").count())
```

```python
%%timeit -n5 -r10

joined_df = raw_df.merge(raw_df, on=['country_of_birth', 'date_of_birth'], how='inner')
diff_df = joined_df[joined_df['user_id_x'] != joined_df['user_id_y']]
diff_df['diff'] = abs(diff_df['score_x']-diff_df['score_y'])
diff_df.groupby('country_of_birth').mean().sort_values(by='diff', ascending=False)
```
